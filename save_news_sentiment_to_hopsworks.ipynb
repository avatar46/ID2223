```python
from datetime import datetime
```

```python
def timestamp_2_time(x):
    
    dt_obj = datetime.strptime(str(x), '%Y-%m-%d %H:%M:%S')
    dt_obj = dt_obj.timestamp() * 1000
    return int(dt_obj)
```

```python
articles_df.publish_date = articles_df.publish_date.apply(timestamp_2_time)
```
