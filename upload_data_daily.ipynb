## Upload news sentiments

```python
apple_articles = sentiment_analysis('Apple', (datetime.today() - timedelta(days=10)))
```

```python
amazon_articles = sentiment_analysis('Amazon', (datetime.today() - timedelta(days=10)))
```

```python
meta_articles = sentiment_analysis('Meta', (datetime.today() - timedelta(days=2)))
```

```python
def timestamp_2_time(x):
    dt_obj = datetime.strptime(str(x), '%Y-%m-%d')
    dt_obj = dt_obj.timestamp() * 1000
    return int(dt_obj)
```

```python
frames = [meta_articles, apple_articles, amazon_articles]
articles_df = pd.concat(frames)
articles_df['date'] = articles_df['date'].apply(lambda x: x.strftime('%Y-%m-%d'))
articles_df.date = articles_df.date.apply(timestamp_2_time)
articles_df
```
