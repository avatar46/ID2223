## <span style="color:#ff5f27;"> 🖍 Create Feature View and Training Data </span>

```python
import hopsworks

project = hopsworks.login() 

fs = project.get_feature_store() 
```

```python
market_sentiments_fg = fs.get_or_create_feature_group(
    name = 'market_news_fg_for_three',
    version = 1
)

stocks_fg = fs.get_or_create_feature_group(
    name = 'stocks_fg',
    version = 1
)
```

```python
query = market_sentiments_fg.select_all()
```

```python
market_sentiments = query.read()
```

```python
query = stocks_fg.select_all()
```

```python
stocks = query.read()
```

```python
market_sentiments
```

```python
keep_columns = ['ticker', 'publish_date', 'neg', 'neu', 'pos', 'compound']
sentiment_df = market_sentiments[keep_columns]
```

```python
import pandas as pd
from datetime import datetime, timedelta

def time_2_datetime(x):
    
    dt_obj = datetime.fromtimestamp(x / 1000)
    return dt_obj

def datetime_2_int(x):
    
    dt_obj = datetime.strptime(str(x), '%Y-%m-%d')
    dt_obj = dt_obj.timestamp() * 1000
    return int(dt_obj)
```

```python
sentiment_df.publish_date = sentiment_df.publish_date.apply(time_2_datetime)
sentiment_df.sort_values(by='publish_date')
```

```python
acc_senti_df = sentiment_df.groupby([sentiment_df['publish_date'].dt.date, 'ticker']).agg({'neg': 'mean', 'neu': 'mean', 'pos': 'mean', 'compound': 'mean'}).reset_index()

acc_senti_df
```

```python
acc_senti_df.publish_date = acc_senti_df.publish_date.apply(datetime_2_int)
acc_senti_df = acc_senti_df.rename(columns={'publish_date': 'date', 'ticker': 'name'})
acc_senti_df['name'] = acc_senti_df['name'].str.upper()
acc_senti_df
```

```python
stocks_with_sentiment = pd.merge(acc_senti_df, stocks, on=['date', 'name'], how='right')
stocks_with_sentiment = stocks_with_sentiment.fillna(0)
```

```python
missing = stocks_with_sentiment.isnull().sum().sort_values(ascending = False)
missing.to_frame(name='total').style.background_gradient()
```

```python
feature_group = fs.get_or_create_feature_group(
        name = 'final_data_for_prediction',
        description = 'Combined data for Apple, Amazon, and Meta\' prediction',
        version = 1,
        primary_key = ['date','name'],
        online_enabled = True,
        event_time = 'date'
    )    
feature_group.insert(stocks_with_sentiment)
```

```python
query = feature_group.select_all()
```

```python
feature_view = fs.create_feature_view(
    name = 'stock_prediction_fv_combined',
    version = 1,
    transformation_functions = {},
    query = query
)
```

```python
feature_view = fs.get_feature_view(
    name = 'stock_prediction_fv_combined',
    version = 1
)
```
