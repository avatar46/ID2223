# Vader Scores

```python
pip install vaderSentiment
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
# init the sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

def predicted_label(x):
  if x<=-0.5:
    return 0
  elif x>=0.5:
    return 2
  else:
    return 1

def score_Vader(df,analyzer):
  df['neg'] = df['body_text'].apply(lambda x:analyzer.polarity_scores(x)['neg'])
  df['neu'] = df['body_text'].apply(lambda x:analyzer.polarity_scores(x)['neu'])
  df['pos'] = df['body_text'].apply(lambda x:analyzer.polarity_scores(x)['pos'])
  df['compound'] = df['body_text'].apply(lambda x:analyzer.polarity_scores(x)['compound'])
  
  df['predicted_class'] = df['compound'].map(predicted_label) 
  return df
```

```python
apple_withScore=score_Vader(apple_data, analyzer)
amazon_withScore=score_Vader(amazon_data, analyzer)
meta_withScore=score_Vader(meta_data, analyzer)
```

```python
apple_withScore
```

```python
apple_withScore.to_pickle("apple_news_sentiments.pkl")
amazon_withScore.to_pickle("amazon_news_sentiments.pkl")
meta_withScore.to_pickle("meta_news_sentiments.pkl")
```

# Aggregate scores by day

```python
keep_columns = ['ticker', 'publish_date', 'neg', 'neu', 'pos', 'compound']
apple_sentiment_df = apple_withScore[keep_columns]
amazon_sentiment_df = amazon_withScore[keep_columns]
meta_sentiment_df = meta_withScore[keep_columns]


```

```python
print(apple_sentiment_df.shape)
print(amazon_sentiment_df.shape)
print(meta_sentiment_df.shape)
```

```python
import pandas as pd

sentiment_df=pd.concat([apple_sentiment_df,amazon_sentiment_df,meta_sentiment_df])
```

```python
print(sentiment_df.shape)
```

```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
sentiment_df['publish_date'] = pd.to_datetime(sentiment_df['publish_date'], format='%b %d, %Y %I:%M%p ET')
```

```python
sentiment_df.sort_values(by='publish_date')
```

```python
acc_senti_df = sentiment_df.groupby([sentiment_df['publish_date'].dt.date, 'ticker']).agg({'neg': 'mean', 'neu': 'mean', 'pos': 'mean', 'compound': 'mean'}).reset_index()

acc_senti_df
```

```python
def datetime_2_int(x):
    
    dt_obj = datetime.strptime(str(x), '%Y-%m-%d')
    dt_obj = dt_obj.timestamp() * 1000
    return int(dt_obj)
```

```python
acc_senti_df.publish_date = acc_senti_df.publish_date.apply(datetime_2_int)
acc_senti_df = acc_senti_df.rename(columns={'publish_date': 'date', 'ticker': 'name'})
acc_senti_df['name'] = acc_senti_df['name'].str.upper()
acc_senti_df
```
